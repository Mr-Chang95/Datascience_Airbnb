### **Neighbourhood Distribution in Seattle and Boston:**

```python
plt.rcParams['figure.figsize'] = [16, 10] #set figure size

seattle_neighbourhood = seattle_listings['neighbourhood_cleansed'].value_counts()
(seattle_neighbourhood/seattle_listings.shape[0]).plot(kind = 'bar')
plt.title('Seattle Neighbourhood Distribution: Proportion of Rented Houses')
plt.savefig('Seattle Neighbourhood Distribution.png');
```

```python
plt.rcParams['figure.figsize'] = [16, 10] #set figure size

boston_neighbourhood = boston_listings['neighbourhood_cleansed'].value_counts()
(boston_neighbourhood/boston_listings.shape[0]).plot(kind = 'bar')
plt.title('Boston Neighbourhood Distribution: Proportion of Rented Houses')
plt.savefig('Boston Neighbourhood Distribution.png');
```

**Neighbourhood Notes:**
> **`1.` There are a greater number of neighbourhoods in Seattle that rent out properties than Boston. This could simply be that there are far more neighbourhood in Seattle than Boston.** <br>
**`2.` For Seattle, the top neighborhood(Broadway) is the only one that takes up a double-digit percentage. The next top neighbourhood is only half of Broadway's proportion.**

### **Boxplot: Prices Range By Neighbourhood**

```python
boston_listings['price'].sort_values() #need to clean up the values before we can plot and analyze
```

```python
plt.rcParams['figure.figsize'] = [16, 10] #set figure size
#Get rid of "," and convert to 'price' column to integer
boston_listings['int_price'] = boston_listings['price'].map(lambda x: int(x[1:-3].replace(",","")))

sb.boxplot(x = 'int_price', y = 'neighbourhood_cleansed', data = boston_listings)
plt.title("Boxplot Distribution of Prices By Boston Neighbourhood");
```

```python
plt.rcParams['figure.figsize'] = [15, 18] #set figure size
seattle_listings['int_price'] = seattle_listings['price'].map(lambda x: int(x[1:-3].replace(",","")))
sb.boxplot(x = 'int_price', y = 'neighbourhood_group_cleansed', data = seattle_listings)
plt.title("Boxplot Distribution of Prices By Seattle Neighbourhood");
```

**Neighbourhood and Prices Notes:**
> **`1.` There are quite a number of outliers in both datasets. However, when comparing the 2 cities visually, we can tell that Seattle neighbourhoods' rentals are more expensive on average, along with a much wider range(IQR).** <br>
**`2.` However, it's a little hard to examine because of the many outliers, so I will be trimming the outliers to get a better look. I will only be interested in those less than or equal to the largest overall max of the IQR of each city. Rather than using the `describe()` function, we will derive it visually.**

```python
plt.rcParams['figure.figsize'] = [16, 10] #set figure size
#Get rid of "," and convert to 'price' column to integer
boston_listings['int_price'] = boston_listings['price'].map(lambda x: int(x[1:-3].replace(",","")))

sb.boxplot(x = 'int_price', y = 'neighbourhood_cleansed',
           data = boston_listings[boston_listings['int_price'] <= 500])
plt.title("Boxplot Distribution of Prices By Boston Neighbourhood: 500 Or Less")
plt.savefig('Boxplot Distribution of Prices By Boston Neighbourhood.png');
```

```python
plt.rcParams['figure.figsize'] = [15, 18] #set figure size
#Get rid of "," and convert to 'price' column to integer
seattle_listings['int_price'] = seattle_listings['price'].map(lambda x: int(x[1:-3].replace(",","")))

sb.boxplot(x = 'int_price', y = 'neighbourhood_group_cleansed',
           data = seattle_listings[seattle_listings['int_price'] <= 500])
plt.title("Boxplot Distribution of Prices By Seattle Neighbourhood: 500 Or Less")
plt.savefig('Boxplot Distribution of Prices By Seattle Neighbourhood.png');
```

**More Neighbourhood and Prices Notes:**
>**`1.` Now that we have a better look, we can see that the Leather District neighbourhood has the highest mean price for the city of Boston. For the city of Seattle, we can see that the mean prices are generally cheaper and are much closer to one another. Downtown Seattle takes the cake for highest mean price.**

### Lineplot Of Neighbourhoods Average Ratings

**Seattle**

```python
plt.rcParams['figure.figsize'] = [16, 10] #set figure size
neighbourhood_ratings = seattle_listings.groupby('neighbourhood_group_cleansed')['review_scores_rating'].mean()
plt.plot(neighbourhood_ratings, color = "black")
plt.xticks(rotation = 80)
plt.title('Lineplot Of Rental Neighbourhoods Average Ratings: Seattle')
plt.savefig('Lineplot Of Rental Neighbourhoods Average Ratings: Seattle.png');
```

**Boston**

```python
plt.rcParams['figure.figsize'] = [16, 10] #set figure size
neighbourhood_ratings = boston_listings.groupby('neighbourhood_cleansed')['review_scores_rating'].mean()
plt.plot(neighbourhood_ratings, color = 'black')
plt.xticks(rotation = 80)
plt.title('Lineplot Of Rental Neighbourhoods Average Ratings: Boston')
plt.savefig('Lineplot Of Rental Neighbourhoods Average Ratings: Boston.png');
```

**Average Rating Notes:**
>**`1.` For the plot we created of average rating of each Seattle neighbourhood, we can see that the average ratings are pretty high, mostly hovering around 93. However, there is a massive drop in rating for the Univeristy District neighbourhood- about 5 points.The highest rating belongs to the Central Area neighbourhood.** <br>
**`2.` For the Boston lineplot, we can see that most of the averges are somewhat lower - hovering between 90 to 94. The highest rating belongs to the Leather District which also happens to have the highest average price. The lowest rating belongs to Mission Hill.**
